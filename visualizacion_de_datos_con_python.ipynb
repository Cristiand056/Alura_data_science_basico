# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df.head()
```

```python
anios = list(map(str, range(1980, 2014)))
anios
```

```python
colombia = df.loc['Colombia', anios]
colombia
```

```python
col_dict = {'Año': colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.title('Inmigración de colombianos hacia Canadá')
plt.xlabel('Año')
plt.ylabel('Múmero de Inmigrantes')
plt.show()
```

##**Desafio 1**

```python
brasil = df.loc['Brasil', anios]
brasil
```

```python
argentina = df.loc['Argentina', anios]
argentina
```

```python
bra_dict = {'Año': brasil.index.tolist(),
            'Inmigrantes': brasil.values.tolist()}
bra_dict
```

```python
arg_dict = {'Año': argentina.index.tolist(),
            'Inmigrantes': argentina.values.tolist()}
arg_dict
```

```python
datos_brasil = pd.DataFrame(bra_dict)
datos_brasil.tail()
```

```python
datos_argentina = pd.DataFrame(arg_dict)
datos_argentina.tail()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_brasil['Año'], datos_brasil['Inmigrantes'])
plt.plot(datos_argentina['Año'], datos_argentina['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.title('comparación de inmigración brasileños y argentinos, hacia Canadá')
plt.xlabel('Año')
plt.ylabel('Múmero de Inmigrantes')
plt.legend(['Brasil', 'Argentina'])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de colombianos hacia Canadá\n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('Múmero de Inmigrantes')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])
axs[0].set_title('Inmigración de colombianos hacia Canadá\n1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Múmero de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid(True)

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la Inmigración de colombianos\nhacia Canadá 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Múmero de Inmigrantes')
axs[1].grid(True)

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Inmigración de los 4 mayores paises de sur america\n hacia Canadá desde 1980 - 2013 hasta')

axs[0,0].plot(df.loc['Colombia', anios])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil', anios])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina', anios])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú', anios])
axs[1,1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Múmero de Inmigrantes')
  ax.grid()
plt.show()
```

##**Desafio 2**

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

```python
meses = list(ventas_2022.keys())
meses
```

```python
ventas_2022_df = pd.DataFrame(ventas_2022, index=tiendas)
ventas_2022_df
```

```python
tienda_A = ventas_2022_df.loc['A', meses]
tienda_A
```

```python
tienda_B = ventas_2022_df.loc['B', meses]
tienda_C = ventas_2022_df.loc['C', meses]
tienda_D = ventas_2022_df.loc['D', meses]
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Ventas de las 4 tiendas durante el año 2022')

axs[0,0].plot(tienda_A)
axs[0,0].set_title('Tienda A')

axs[0,1].plot(tienda_B)
axs[0,1].set_title('Tienda B')

axs[1,0].plot(tienda_C)
axs[1,0].set_title('Tienda C')

axs[1,1].plot(tienda_D)
axs[1,1].set_title('Tienda D')

ymin = 0
ymax = 500

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

for ax in axs.flat:
  #ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Mes')
  ax.set_ylabel('Número de Ventas')
  ax.grid()
plt.show()
```
