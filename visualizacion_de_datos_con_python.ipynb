# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df.head()
```

```python
anios = list(map(str, range(1980, 2014)))
anios
```

```python
colombia = df.loc['Colombia', anios]
colombia
```

```python
col_dict = {'Año': colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.title('Inmigración de colombianos hacia Canadá')
plt.xlabel('Año')
plt.ylabel('Múmero de Inmigrantes')
plt.show()
```

##**Desafio 1**

```python
brasil = df.loc['Brasil', anios]
brasil
```

```python
argentina = df.loc['Argentina', anios]
argentina
```

```python
bra_dict = {'Año': brasil.index.tolist(),
            'Inmigrantes': brasil.values.tolist()}
bra_dict
```

```python
arg_dict = {'Año': argentina.index.tolist(),
            'Inmigrantes': argentina.values.tolist()}
arg_dict
```

```python
datos_brasil = pd.DataFrame(bra_dict)
datos_brasil.tail()
```

```python
datos_argentina = pd.DataFrame(arg_dict)
datos_argentina.tail()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_brasil['Año'], datos_brasil['Inmigrantes'])
plt.plot(datos_argentina['Año'], datos_argentina['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.title('comparación de inmigración brasileños y argentinos, hacia Canadá')
plt.xlabel('Año')
plt.ylabel('Múmero de Inmigrantes')
plt.legend(['Brasil', 'Argentina'])
plt.show()
```
